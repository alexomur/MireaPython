```python
# 4.1

import numpy as np
import matplotlib.pyplot as plt
from math import floor, ceil

def generate_NxM(N: int = 5, M: int = 5, color: bool = False):
    if color:
        mat = np.random.randint(0, 256, size=(M, ceil(N / 2), 3), dtype=np.uint8)
    else:
        mat = np.random.choice(2, M * ceil(N / 2)).reshape(M, ceil(N / 2))
    sym_part = np.fliplr(mat[:, :floor(N / 2)])
    extended_mat = np.concatenate((mat, sym_part), axis=1)
    return extended_mat

plt.imshow(generate_NxM(), cmap='gray')
plt.show()
```

```python
# 4.2

def generate_map(a: int = 10, b: int = 20, N: int = 5, M: int = 5, color = False):
    map_height = a * (M + 5)
    map_width = b * (N + 5)
    if color:
        mp = 255 * np.ones((map_height, map_width, 3), dtype=np.uint8)
    else:
        mp = np.zeros((a * (N + 5), b * (M + 5)))
    
    for i in range(a):
        for j in range(b):
            pic = generate_NxM(N, M, color)
            start_row = i * (M + 5)
            start_col = j * (N + 5)
            
            mp[start_row:start_row+M, start_col:start_col+N] = pic
    return mp

# 4.2 - False
# 4.3 - True
mp = generate_map(color = True)

print(mp)

plt.imshow(mp)
plt.axis('off')
plt.show()
```
